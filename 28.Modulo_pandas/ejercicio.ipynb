<h3>Empezamos a procesar la url que contiene mucha informacion</h3>

```python
# Me quedo con las url que no son nulas
serie_url_landing = df_navegacion["url_landing"][df_navegacion["url_landing"].notnull()]
# Con la url, utilizamos split=& y cada fragmaneto genero una nueva columna
df_navegacion[["url_base", "idUser", "uuid_url", "camp", "adg", "device", "sl", "adv", "rec","vacia"]]=(serie_url_landing.str.split(pat="&", n=10, expand=True))
```

```python
# Eliminar las columnas que no necesitamos: idUser, uuid_url, rec, vacia axis=1
df_navegacion = df_navegacion.drop(['idUser', 'uuid_url', 'rec', 'vacia'], axis=1)
```

```python
# Comprobar url_base url/modelo?parametros
df_navegacion['url_base']
# split por ? lambda

df_navegacion['url_base'] = df_navegacion['url_base'].apply(lambda url: str(url).split('?')[0])
df_navegacion['url_base']

```

```python
# buscar la posicion de la ultima /    funcion obtenerModelo
# nos quedamos con el modelo y generamos columna modelo 
def obtenerModelo(url):
    posicion = url.rfind('/')
    modelo = url[posicion + 1:]
    if modelo == '' or modelo == 'renting':
        modelo = 'home'
    return modelo
    
df_navegacion.insert(loc=7, column='modelo', value= df_navegacion['url_base'].apply(obtenerModelo))    
print(df_navegacion['modelo'].head(50))
```

### Crear un orden de usuario, Como tenemos usuarios repetidos me daria un problema al haver el merge con conversiones,csv

```python
# Crear una columna llamada orden que almacene el orden de los id_user
df_navegacion = df_navegacion.sort_values(['id_user','ts'])
df_navegacion.insert(loc=1, column='orden', value=df_navegacion.groupby('id_user')['ts'].cumcount())
```

```python
# Filtrar por orden  = 0

df_navegacion = df_navegacion[df_navegacion['orden']==0]
df_navegacion['orden'].head(50)
```

### Empezamos a procesar conversiones.csv

```python
# Craer df_conversiones

df_conversiones = pd.read_csv('archivos_practica/conversiones.csv', sep=';')
df_conversiones
```

```python
# Hacer el  merge por id_user con modo outer para no perder nada de informacion
df_final = pd.merge(df_navegacion, df_conversiones, how='outer', on=['id_user'])
df_final
```
